```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.metrics import accuracy_score, classification_report
from xgboost import XGBClassifier
```

```python
df = pd.read_csv(file2)
```

```python
len(df.columns)
```

```python
X = df.drop('prognosis', axis=1)
y = df['prognosis']
```

```python
y.value_counts()
```

#### Encode string labels to integers

```python
le = LabelEncoder()
y_encoded = le.fit_transform(y)
```

```python
y_encoded
```

```python
list(le.classes_)
```

```python
le.inverse_transform([3,5])
```

```python
X_train, X_test, y_train, y_test = train_test_split(
    X, y_encoded, test_size=0.2, random_state=42, stratify=y_encoded)
```

### Create model

```python
model = XGBClassifier(
    objective='multi:softprob',
    num_class=len(le.classes_), 
    eval_metric='mlogloss',
    use_label_encoder=False,
    max_depth=4,
    learning_rate=0.3,
    n_estimators=100,
    random_state=42
)

model.fit(X_train, y_train)
```

```python
# Predict class labels
y_pred = model.predict(X_test)

# Predict class probabilities
y_prob = model.predict_proba(X_test)


print("\n Accuracy:", accuracy_score(y_test, y_pred))
print("\n Classification Report:\n",
      classification_report(y_test, y_pred, target_names=le.classes_))

print("\n Predicted class labels:", y_pred[:5])
print(" Actual class labels   :", y_test[:5])

print("\n Predicted probabilities for first 5 samples:\n", y_prob[1])
```

```python
def cosine_similarity(vec1, vec2):
    dot_product = np.dot(vec1, vec2)
    norm_vec1 = np.linalg.norm(vec1)
    norm_vec2 = np.linalg.norm(vec2)
    return dot_product / (norm_vec1 * norm_vec2)
```

```python
def get_symptoms(symptoms_list):
    results = []
    idx = []
    for symptom in symptoms_list:
        em_sym = hf.embed_query(symptom)
        sim_res = cosine_similarity(embeded_symproms, em_sym)
        id = sim_res.argmax()
        idx.append(id)
        results.append(symptoms[id])
    return results, idx
```

```python
from sentence_transformers import SentenceTransformer, util
import numpy as np

# Модель embedding (можно твою hf, но SentenceTransformer лучше для этого)
model = SentenceTransformer(
    'all-MiniLM-L6-v2',
    device="cpu"
)


new_symptoms = list(df.columns)
old_symptoms = list(df_2.columns)

new_emb = model.encode(new_symptoms, convert_to_tensor=True)

# Mapping
mapping = {}
for old_s in old_symptoms:
    old_emb = model.encode(old_s, convert_to_tensor=True)
    sim = util.cos_sim(old_emb, new_emb)[0]
    best_idx = np.argmax(sim)
    best_new = new_symptoms[best_idx]
    best_sim = sim[best_idx]

    if best_sim > 0.5:  # Порог — ниже не маппим
        mapping[old_s] = best_new
        print(f"Mapped '{old_s}' → '{best_new}' (sim: {best_sim:.2f})")
    else:
        print(f"No good match for '{old_s}' (max sim: {best_sim:.2f}) — will add as new column")


```

```python
old_df.columns = [mapping.get(col, col) if col not in  ["disease","prognosis"] else col for col in old_df.columns]
```
